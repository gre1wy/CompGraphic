```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def recursion(left_limit, right_limit, iter, n):
    if n == iter:
        return 0
    else:
        orig = np.linspace(left_limit, right_limit, 11)

        part_one = orig[0:4]  # от 0 до 0.3
        limits_part_one = part_one[0], part_one[-1]

        part_two = orig[4:8] # от 0.4 до 0.7
        limits_part_two = part_two[0], part_two[-1]  
        
        part_three = orig[8:] # от 0.8 до 1
        limits_part_three = part_three[0], part_three[-1]

        plt.plot([limits_part_one[0], limits_part_one[1]], [-0.02*n, -0.02*n], color='blue')
        plt.plot([limits_part_two[0], limits_part_two[1]], [-0.02*n,-0.02*n], color='blue')  
        plt.plot([limits_part_three[0], limits_part_three[1]], [-0.02*n, -0.02*n], color='blue')

        recursion(limits_part_one[0], limits_part_one[1], iter, n+1)
        recursion(limits_part_two[0], limits_part_two[1], iter, n+1)
        recursion(limits_part_three[0], limits_part_three[1], iter, n+1)
```

```python
def fractal(iterations: int):

    orig = np.linspace(0, 1, 11)

    part_one = orig[0:4]  # от 0 до 0.3
    limits_part_one = part_one[0], part_one[-1]

    part_two = orig[4:8] # от 0.4 до 0.7
    limits_part_two = part_two[0], part_two[-1]  
    
    part_three = orig[8:] # от 0.8 до 1
    limits_part_three = part_three[0], part_three[-1]  

    plt.figure(figsize=(18, 10))

    plt.plot([0, 1], [0.02, 0.02], color='blue')
    plt.plot([limits_part_one[0], limits_part_one[1]], [0, 0], color='blue')
    plt.plot([limits_part_two[0], limits_part_two[1]], [0, 0], color='blue')  
    plt.plot([limits_part_three[0], limits_part_three[1]], [0, 0], color='blue')

    recursion(limits_part_one[0], limits_part_one[1], iterations, 1)
    recursion(limits_part_two[0], limits_part_two[1], iterations, 1)
    recursion(limits_part_three[0], limits_part_three[1], iterations, 1)

    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Psevdo Cantor')
    plt.grid(True)
    plt.show()
```

```python
fractal(4)
```
