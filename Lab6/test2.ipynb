```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

bigger_image = cv2.imread('Images\doggy.jpg')

# Преобразование изображения из BGR в RGB
bigger_image_rgb = cv2.cvtColor(bigger_image, cv2.COLOR_BGR2RGB)

# Отображение изображения с помощью Matplotlib
plt.imshow(bigger_image_rgb)
plt.show()
```
